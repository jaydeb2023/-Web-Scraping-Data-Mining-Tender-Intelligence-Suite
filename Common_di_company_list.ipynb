```python
import pandas as pd

# Read the files
df1 = pd.read_excel('ductile iron pipe_ENSUN_COMPANY_DETAILS.xlsx')
df2 = pd.read_excel('DI_ensun_MANUFACTURE_COMAPNY_DETAILS.xlsx')

# Merge on 'Company Name' with suffixes for clarity
merged = pd.merge(df1, df2, on='Company Name', how='outer', suffixes=('_Sheet1', '_Sheet2'), indicator=True)

# Add Source Column based on merge indicator
merged['Source'] = merged['_merge'].map({'both': 'Common', 'left_only': 'Unique_Sheet1', 'right_only': 'Unique_Sheet2'})

# Drop merge indicator
merged.drop(columns=['_merge'], inplace=True)

# Optional: Rearranging columns to bring 'Source' upfront
cols = merged.columns.tolist()
cols = ['Company Name', 'Source'] + [col for col in cols if col not in ['Company Name', 'Source']]
merged = merged[cols]

# Export to Excel
merged.to_excel('Ensun_Di_Ductile_pipe_Correct_Output.xlsx', index=False)

print("✅ Clean comparison file generated as 'Ensun_Di_Ductile_pipe_Correct_Output.xlsx'")
```
