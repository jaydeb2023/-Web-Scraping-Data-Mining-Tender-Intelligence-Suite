```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Credentials
USERNAME = "dredgingtender@rashmigroup.com"
PASSWORD = "Rashmi@2022"
LOGIN_URL = "https://www.tenderdetail.com/Account/LogOn"

# Launch browser
driver = webdriver.Chrome()
driver.get(LOGIN_URL)
driver.maximize_window()
time.sleep(3)  # Allow page to load

# Updated selectors based on live site inspection
try:
    # Use name attribute instead of ID
    driver.find_element(By.NAME, "UserName").send_keys(USERNAME)
    driver.find_element(By.NAME, "Password").send_keys(PASSWORD)

    print("🔐 Please solve the CAPTCHA manually in the browser, then press Enter here...")
    input()  # Wait for user to solve CAPTCHA manually

    # Click login
    driver.find_element(By.XPATH, '//button[text()="Sign In"]').click()

    # Wait and verify login
    time.sleep(5)
    if "Dashboard" in driver.page_source or "Logout" in driver.page_source:
        print("✅ Login successful, dashboard loaded.")
    else:
        print("❌ Login may have failed. Please check manually.")
except Exception as e:
    print(f"❌ Error: {e}")
```
