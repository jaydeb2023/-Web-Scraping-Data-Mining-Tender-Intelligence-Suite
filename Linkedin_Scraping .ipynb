```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

# Setup driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Manual login
driver.get("https://www.linkedin.com/login")
input("🔐 Please log in manually and press Enter here once done...")

# Visit company page
company_url = "https://www.linkedin.com/company/g-o-r-i-spa/"
driver.get(company_url)
time.sleep(5)

### --- Part 1: Scrape 'About' Information --- ###
try:
    about_tab = driver.find_element(By.XPATH, "//a[contains(@href, '/about')]")
    about_tab.click()
    time.sleep(5)

    print("\n=== 🏢 COMPANY INFO ===")
    name = driver.find_element(By.TAG_NAME, 'h1').text
    print("Company Name:", name)

    about_text = driver.find_element(By.XPATH, "//section[contains(@class,'core-section-container')]//p").text
    print("Description:", about_text)

    labels = driver.find_elements(By.CLASS_NAME, 'org-page-details__definition-term')
    values = driver.find_elements(By.CLASS_NAME, 'org-page-details__definition-text')

    print("\n📊 Company Details:")
    for label, value in zip(labels, values):
        print(f"- {label.text.strip()}: {value.text.strip()}")

except Exception as e:
    print("❌ Failed to scrape company details:", e)

### --- Part 2: Scrape People (Employees) Info --- ###
try:
    people_tab = driver.find_element(By.XPATH, "//a[contains(@href, '/people')]")
    people_tab.click()
    time.sleep(5)

    print("\n=== 👥 EMPLOYEES LIST ===")

    # Scroll to load people
    for _ in range(5):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(2)

    # Scrape visible employee blocks
    people_blocks = driver.find_elements(By.CLASS_NAME, 'artdeco-list__item')
    for idx, person in enumerate(people_blocks, 1):
        try:
            name = person.find_element(By.TAG_NAME, 'span').text
            position = person.find_element(By.CLASS_NAME, 'org-people-profile-card__profile-title').text
            print(f"{idx}. {name} — {position}")
        except:
            continue

except Exception as e:
    print("❌ Failed to scrape employees:", e)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Keywords to filter in title/designation
keywords = [
    "Sales Head", "VP Sales", "CMO", "Country head sales",
    "Business Development", "Sales Director"
]

# Setup Selenium driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Login to LinkedIn manually
driver.get("https://www.linkedin.com/login")
input("🔐 Login manually and press ENTER when fully logged in...")

# Go to GORI spa people tab
driver.get("https://www.linkedin.com/company/g-o-r-i-spa/people/")
time.sleep(5)

# Scroll to load more people
for _ in range(8):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2)

# Grab all employee cards
cards = driver.find_elements(By.XPATH, "//li[contains(@class, 'artdeco-list__item')]")

results = []

for card in cards:
    try:
        text = card.text.strip().split('\n')
        if len(text) >= 2:
            name = text[0]
            position = text[1]
            # Check if any keyword is in the position title
            if any(k.lower() in position.lower() for k in keywords):
                try:
                    profile_link = card.find_element(By.TAG_NAME, "a").get_attribute("href")
                except:
                    profile_link = ""
                results.append({
                    "Name": name,
                    "Designation": position,
                    "Company": "GORI spa",
                    "Email": "",  # Email is not shown here
                    "LinkedIn Profile": profile_link
                })
    except Exception:
        continue

# Save to Excel
df = pd.DataFrame(results).drop_duplicates()
df.to_excel("GORI_spa_Sales_Profiles.xlsx", index=False)

print(f"✅ Extracted {len(df)} filtered employees. Saved to 'GORI_spa_Sales_Profiles.xlsx'")

driver.quit()
```
