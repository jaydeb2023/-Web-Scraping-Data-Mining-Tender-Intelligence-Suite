```python
tender_details_id = 'BOQ_1833226.xls'
```

```python
def extract_class(description):
    classes = ['DI K-9', 'DI K9', 'DI K-7', 'DI K7', 'MS', 'PVC', 'HDPE', 'CI', 'DI PIPE', 'M.S', 'Ductile Iron']
    for c in classes:
        if c.lower() in description.lower():
            return c.upper()
    return ''

def extract_dia(description):
    match = re.search(r'(\d{2,4})\s*mm', description)
    if match:
        return int(match.group(1))
    return ''
```

```python
result = []
for idx, row in df.iterrows():
    item_desc = str(row['Item Description'])
    class_ = extract_class(item_desc)
    dia = extract_dia(item_desc)
    length = row.get('Quantity', '')  
    ref_id = row.get('Estimated Rate', '') 
result.append({
        'Tender Details ID': tender_details_id,
        'Tender/Ref ID': ref_id,
        'Name of Work': item_desc,
        'Class': class_,
        'Diameter (mm)': dia,
        'Length': length
    })
result_df = pd.DataFrame(result)
```

```python
import pandas as pd
import re

# Read BOQ skipping header garbage
df = pd.read_excel("BOQ_1833226.xls", skiprows=10)

# Regex patterns
class_pattern = r"\b(DI|HDPE|PVC|UPVC|CI|MS)\b"
dia_pattern = r"\b(\d{2,4})\s*(mm|MM)?\b"

# Column inference
description_column = next((col for col in df.columns if "description" in col.lower()), None)
qty_column = next((col for col in df.columns if "qty" in col.lower()), None)
unit_column = next((col for col in df.columns if "unit" in col.lower()), None)
rate_column = next((col for col in df.columns if "rate" in col.lower()), None)
amount_column = next((col for col in df.columns if "amount" in col.lower()), None)

if not description_column:
    raise Exception("Description column not found.")

name_of_work = str(df.iloc[0][description_column])[:80]
results = []

for idx, row in df.iterrows():
    text = str(row.get(description_column, "")).strip()

    # Skip junk rows like "200 mm dia Pipe"
    if re.fullmatch(r"\d{2,4}\s*mm\s*dia\s*pipe", text, re.IGNORECASE):
        continue

    # Skip empty
    if not text or text.lower() in ["nan", ""]:
        continue

    class_match = re.findall(class_pattern, text)
    dia_match = re.findall(dia_pattern, text)

    dia_values = [m[0] for m in dia_match]
    class_values = list(set(class_match))

    # Pull unit and quantity info if available
    unit_val = str(row.get(unit_column, "")).strip() if unit_column else ""
    qty_val = row.get(qty_column, "") if qty_column else ""
    rate_val = row.get(rate_column, "") if rate_column else ""
    amount_val = row.get(amount_column, "") if amount_column else ""

    result = {
        "tender_details_id": "BOQ_1833226",
        "name_of_work": name_of_work,
        "item_description": text[:150],
        "class": ",".join(class_values),
        "DIA": ",".join(set(dia_values)),
        "unit": unit_val,
        "quantity": qty_val,
        "estimate rate": rate_val,
        "total amount with taxes": amount_val
    }

    # Only keep rows with real content
    if result["class"] or result["DIA"] or result["quantity"]:
        results.append(result)

# Output
final_df = pd.DataFrame(results)
final_path = "final_cleaned_tender_output.xlsx"
final_df.to_excel(final_path, index=False)

print("✅ File saved:", final_path)
final_df.head()
```

```python
import pandas as pd
import re
import numpy as np

# Load Excel data
df = pd.read_excel("BOQ_1833226.xls", skiprows=10)

# Identify column names
description_column = next((col for col in df.columns if isinstance(col, str) and "description" in col.lower()), None)
qty_column = next((col for col in df.columns if isinstance(col, str) and "qty" in col.lower()), None)
unit_column = next((col for col in df.columns if isinstance(col, str) and "unit" in col.lower()), None)
rate_column = next((col for col in df.columns if isinstance(col, str) and "rate" in col.lower()), None)
amount_column = next((col for col in df.columns if isinstance(col, str) and "amount" in col.lower()), None)

if not description_column:
    raise Exception("Could not find 'Item Description' column.")

# Regex patterns
class_pattern = r"\b(DI|HDPE|PVC|UPVC|CI|MS)\b"
dia_pattern = r"\b(\d{2,4})\s*(mm|MM)?\b"

# Extract name_of_work from first valid row
name_of_work = str(df.iloc[0][description_column])[:80]

results = []

for _, row in df.iterrows():
    text = str(row.get(description_column, "")).strip()

    if len(text) < 5 or text.lower() in ["nan", ""]:
        continue

    class_match = re.findall(class_pattern, text)
    dia_match = re.findall(dia_pattern, text)

    dia_values = [m[0] for m in dia_match]
    class_values = list(set(class_match))

    def safe_number(val):
        try:
            return float(val) if pd.notna(val) else None
        except:
            return None

    unit_val = str(row.get(unit_column, "")).strip() if unit_column else ""
    rate_val = safe_number(row.get(rate_column, "")) if rate_column else None
    amount_val = safe_number(row.get(amount_column, "")) if amount_column else None
    qty_val = safe_number(row.get(qty_column, "")) if qty_column else None

    # Calculate quantity if it's missing
    if qty_val in [None, "", 0] and rate_val and amount_val:
        try:
            qty_val = round(amount_val / rate_val, 2)
        except ZeroDivisionError:
            qty_val = ""

    result = {
        "tender_details_id": "BOQ_1833226",
        "name_of_work": name_of_work,
        "item_description": text[:150],
        "class": ",".join(class_values),
        "DIA": ",".join(set(dia_values)),
        "unit": unit_val,
        "quantity": qty_val,
        "estimate rate": rate_val,
        "total amount with taxes": amount_val
    }

    if result["class"] or result["DIA"] or result["quantity"]:
        results.append(result)

# Create and save final DataFrame
final_df = pd.DataFrame(results)
final_df.replace({np.nan: ""}, inplace=True)

output_path = "boq_final_with_quantity.xlsx"
final_df.to_excel(output_path, index=False)

print(f" File saved: {output_path}")
final_df.head()
```
