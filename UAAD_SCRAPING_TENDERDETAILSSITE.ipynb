```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
import time

# ✅ Update this path if ChromeDriver location changes
driver_path = r"C:\Users\Jaydeb\.cache\selenium\chromedriver\win64\137.0.7151.68\chromedriver.exe"

# 🔧 Browser setup
options = Options()
options.add_argument("--start-maximized")
# options.add_argument("--headless")  # Uncomment to run in background

# 🚀 Launch browser
driver = webdriver.Chrome(service=Service(driver_path), options=options)

# 🌐 Target URL
url = "https://www.tenderdetail.com/government-tenders/urban-administration-and-development-tenders/1?agid=3726"
driver.get(url)
time.sleep(5)  # Allow full page load

# 📜 Scroll to bottom to trigger lazy loading if needed
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)

# 📦 Extract tender blocks
tenders = driver.find_elements(By.CLASS_NAME, "tender_row")

data = []

# 🔍 Extract fields from each tender block
for tender in tenders:
    try:
        tender_id = tender.find_element(By.CLASS_NAME, "m-tender-id").text.strip()
        title = tender.find_element(By.CLASS_NAME, "m-brief").text.strip()
        link = tender.find_element(By.CLASS_NAME, "m-brief").get_attribute("href")

        # 🆕 Use combined block text to extract due date and value
        full_info_text = tender.text

        # Extract Due Date
        due_date = "N/A"
        if "Due Date :" in full_info_text:
            due_date = full_info_text.split("Due Date :")[1].split("Tender Value")[0].strip()

        # Extract Tender Value
        tender_value = "N/A"
        if "Tender Value :" in full_info_text:
            tender_value = full_info_text.split("Tender Value :")[1].split("View Notice")[0].strip()

        data.append({
            "Tender ID": tender_id,
            "Title": title,
            "Due Date": due_date,
            "Tender Value": tender_value,
            "Detail Link": link
        })
    except Exception as e:
        print("❌ Error parsing a tender:", e)

# 🛑 Close browser
driver.quit()

# 💾 Save data to Excel
df = pd.DataFrame(data)
df.to_excel("UAD_Tenders_Live.xlsx", index=False)

print("✅ Done! Data saved to 'UAD_Tenders_Live.xlsx'")
```
