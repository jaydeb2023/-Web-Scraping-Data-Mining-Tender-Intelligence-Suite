```python
import pandas as pd

# Load the Excel file
file_path = "all_data.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet1")

# Function to extract pipe class/type
def extract_class(text):
    text = str(text).lower()
    if 'k-7' in text:
        return 'K-7'
    elif 'k-9' in text or 'k9' in text:
        return 'K-9'
    elif 'm.s' in text or 'ms' in text:
        return 'M.S'
    elif 'hdpe' in text:
        return 'HDPE'
    elif 'upvc' in text:
        return 'UPVC'
    elif 'pvc' in text:
        return 'PVC'
    elif 'ci' in text:
        return 'CI'
    elif 'di' in text:
        return 'DI'
    else:
        return None

# Function to extract unit
def extract_unit(text):
    text = str(text).lower()
    if 'mtr' in text or 'meter' in text:
        return 'Mtr'
    elif 'each' in text:
        return 'Each'
    elif 'kg' in text:
        return 'Kg'
    elif 'litre' in text or 'ltr' in text:
        return 'Litre'
    elif 'set' in text:
        return 'Set'
    elif 'nos' in text or 'no.' in text:
        return 'Nos'
    else:
        return None

# Apply to full DataFrame, not filtered by "pipe"
df['Pipe Class'] = df['Tender Brief'].apply(extract_class)
df['Unit'] = df['Tender Brief'].apply(extract_unit)

# Filter rows where either Pipe Class or Unit was found
pipe_df = df[df['Pipe Class'].notnull() | df['Unit'].notnull()]

# Show shape and sample
print(pipe_df.shape)
pipe_df.head()
```

```python
pipe_df.to_excel("tenders_with_class_or_unit.xlsx", index=False)
```
