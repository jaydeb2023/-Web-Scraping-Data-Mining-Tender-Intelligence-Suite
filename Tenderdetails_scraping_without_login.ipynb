```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import time
import pandas as pd

# Initialize driver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=options)
driver.get("https://www.tenderdetail.com/Indian-Tenders")
time.sleep(3)

# List to hold data
tenders = []

# Function to scrape all tenders from current page
def scrape_current_page():
    rows = driver.find_elements(By.CSS_SELECTOR, ".tender_row")
    for row in rows:
        try:
            tdr_id = row.find_element(By.CLASS_NAME, "m-tender-id").text.strip()
            title = row.find_element(By.CSS_SELECTOR, "a.m-brief").get_attribute("title").strip()
            url = row.find_element(By.CSS_SELECTOR, "a.m-brief").get_attribute("href").strip()
            due_date = row.find_element(By.CLASS_NAME, "m-due-date").find_element(By.XPATH, "..").text.replace("Due Date : ", "").strip()
            value = row.find_element(By.CLASS_NAME, "m-value").find_element(By.XPATH, "..").text.replace("Tender Value : ", "").strip()
            tenders.append({
                "TDR": tdr_id,
                "Title": title,
                "URL": url,
                "Due Date": due_date,
                "Tender Value": value
            })
        except Exception as e:
            print(f"⚠️ Skipped one tender due to: {e}")

# Loop through all pages
page = 1
while True:
    print(f"📄 Scraping page {page}...")
    scrape_current_page()

    # Click next page
    try:
        next_btn = driver.find_element(By.LINK_TEXT, "»")
        driver.execute_script("arguments[0].click();", next_btn)
        time.sleep(3)
        page += 1
    except NoSuchElementException:
        print("✅ Reached the last page.")
        break
    except Exception as e:
        print(f"❌ Error navigating to next page: {e}")
        break

# Save data
df = pd.DataFrame(tenders)
df.to_excel("all_indian_tenders.xlsx", index=False)
print(f"\n✅ Scraping complete. Saved {len(tenders)} tenders to 'all_indian_tenders.xlsx'")

driver.quit()
```
