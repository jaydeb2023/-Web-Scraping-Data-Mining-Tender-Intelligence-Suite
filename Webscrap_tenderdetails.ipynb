```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Setup Chrome WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

# Visit Indian Tenders page
driver.get("https://www.tenderdetail.com/Indian-Tenders")

# Wait for tenders to load
WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, "tender_row")))

all_tenders = []

# You can scrape multiple pages (e.g. first 3 pages)
for page in range(1, 4):
    print(f"Scraping page {page}...")
    driver.get(f"https://www.tenderdetail.com/Indian-Tenders/{page}")
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "tender_row")))
    time.sleep(2)  # ensure content is rendered

    rows = driver.find_elements(By.CLASS_NAME, "tender_row")

    for row in rows:
        try:
            org = row.find_element(By.TAG_NAME, "strong").text.strip()
            desc = row.find_element(By.CLASS_NAME, "m-brief").text.strip()
            tid = row.find_element(By.CLASS_NAME, "m-tender-id").text.strip()
            due_date = row.find_element(By.CLASS_NAME, "m-due-date").find_element(By.XPATH, "..").text.strip().replace("Due Date :", "").strip()
            value = row.find_element(By.CLASS_NAME, "tender-value").text.strip()
            link = row.find_element(By.CLASS_NAME, "viewnotice").get_attribute("href")

            all_tenders.append({
                "Tender ID": tid,
                "Organization": org,
                "Description": desc,
                "Due Date": due_date,
                "Tender Value": value,
                "Notice URL": link
            })
        except Exception as e:
            print("⚠️ Error in one tender:", e)

driver.quit()

# Save to CSV
df = pd.DataFrame(all_tenders)
df.to_csv("tender_detail_results.csv", index=False)
print("✅ Saved to 'tender_detail_results.csv'")

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException
import pandas as pd
import time
import random

# Setup Chrome WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

# Data container
all_tenders = []

start_page = 1
end_page = 500  # You can increase this later. 26392 is huge and should be done in chunks.

for page in range(start_page, end_page + 1):
    url = f"https://www.tenderdetail.com/Indian-Tenders/{page}"
    print(f"Scraping {url}")

    for attempt in range(3):  # Retry up to 3 times
        try:
            driver.set_page_load_timeout(60)
            driver.get(url)
            WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "tender_row")))
            break
        except TimeoutException:
            print(f"⚠️ Timeout on {url}, retrying ({attempt+1}/3)...")
            time.sleep(5)
        except Exception as e:
            print(f"❌ Failed to load page {url} due to error: {e}")
            break
    else:
        print(f"❌ Skipping {url} after 3 failed attempts.")
        continue

    time.sleep(random.uniform(2, 4))  # Delay to avoid blocking

    rows = driver.find_elements(By.CLASS_NAME, "tender_row")

    for row in rows:
        try:
            org = row.find_element(By.TAG_NAME, "strong").text.strip()
            desc = row.find_element(By.CLASS_NAME, "m-brief").text.strip()
            tid = row.find_element(By.CLASS_NAME, "m-tender-id").text.strip()
            due = row.find_element(By.CLASS_NAME, "m-due-date").find_element(By.XPATH, "..").text.replace("Due Date :", "").strip()
            value = row.find_element(By.CLASS_NAME, "tender-value").text.strip()
            link = row.find_element(By.CLASS_NAME, "viewnotice").get_attribute("href")

            all_tenders.append({
                "Tender ID": tid,
                "Organization": org,
                "Description": desc,
                "Due Date": due,
                "Value": value,
                "Notice URL": link
            })
        except Exception as e:
            print("⚠️ Skipped a row due to error:", e)

# Cleanup
driver.quit()

# Save data
df = pd.DataFrame(all_tenders)
df.to_csv("tenders_scraped.csv", index=False)
print("✅ Data saved to tenders_scraped.csv")
```
