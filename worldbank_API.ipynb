```python
import requests
import pandas as pd

# API endpoint and parameters
url = "https://search.worldbank.org/api/v3/projects"
params = {
    "format": "json",
    "proj_id": "P180886"
}

# Send request
resp = requests.get(url, params=params)
resp.raise_for_status()
data = resp.json()

# Extract the project object from returned dict
projects = data.get("projects", {})
proj = projects.get("P180886")
if not proj:
    raise ValueError("Project P180886 not found")

# Convert to DataFrame (single-row)
df = pd.json_normalize(proj)

# Save to Excel or CSV
df.to_excel("P180886_Project_Details.xlsx", index=False)
print("✅ Project details saved to P180886_Project_Details.xlsx")
```

```python
import requests
import pandas as pd

# Base URL for Projects API
base_url = "https://search.worldbank.org/api/v3/projects"
params = {
    "format": "json",
    "rows": 100,     # number of records per request (max 100)
    "os": 0          # starting offset
}

all_projects = []
total = None

while True:
    response = requests.get(base_url, params=params)
    response.raise_for_status()
    data = response.json()
    
    # Inspect header details
    if total is None:
        total = int(data.get("total", 0))
        print(f"🔄 Total projects to fetch: {total}")
    
    projects = data.get("projects", {})
    if not projects:
        break
    
    all_projects.extend(projects.values())
    print(f"Fetched {len(all_projects)} / {total}")
    
    # Check if we've retrieved all records
    if len(all_projects) >= total:
        break
    
    params["os"] += params["rows"]  # advance to next batch

# Convert to DataFrame
df = pd.json_normalize(all_projects)

# Save to Excel or CSV
df.to_excel("All_WorldBank_Projects.xlsx", index=False)
print(f"✅ Completed! Saved {len(df)} records.")
```
