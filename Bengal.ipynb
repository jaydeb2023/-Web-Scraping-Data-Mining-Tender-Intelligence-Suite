```python
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# CONFIG
USERNAME = "dredgingtender@rashmigroup.com"
PASSWORD = "Rashmi@2022"
TDR_LIST_FILE = "tdrs.xlsx"  # Column: TDR (e.g., 49454636)
DOWNLOAD_DIR = os.path.join(os.getcwd(), "Downloads")

# Setup Chrome with download behavior
options = webdriver.ChromeOptions()
prefs = {
    "download.default_directory": DOWNLOAD_DIR,
    "download.prompt_for_download": False,
    "plugins.always_open_pdf_externally": True
}
options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Ensure download folder exists
os.makedirs(DOWNLOAD_DIR, exist_ok=True)

def login():
    driver.get("https://www.tenderdetail.com/Account/LogOn")
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "UserName"))).send_keys(USERNAME)
    driver.find_element(By.ID, "Password").send_keys(PASSWORD)
    driver.find_element(By.XPATH, "//input[@type='submit']").click()
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "navbar-custom-menu")))
    print("✅ Logged in successfully.")

def search_tdr(tdr_number):
    search_url = f"https://www.tenderdetail.com/registeruser/searchtenders?ServiceType=1&nm={tdr_number}"
    driver.get(search_url)
    print(f"🔍 Searching TDR: {tdr_number}")
    time.sleep(4)

def download_boq(tdr_number):
    try:
        view_notice = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//a[contains(@class, 'viewnotice')]"))
        )
        view_notice.click()
        time.sleep(4)

        # Switch to new tab
        driver.switch_to.window(driver.window_handles[1])

        # Find and click BOQ download link
        boq_links = driver.find_elements(By.PARTIAL_LINK_TEXT, "BOQ")
        if not boq_links:
            print(f"❌ BOQ not found for TDR {tdr_number}")
        else:
            for link in boq_links:
                print(f"⬇️ Downloading BOQ for {tdr_number}")
                link.click()
                time.sleep(5)

        # Close tab
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

    except Exception as e:
        print(f"❌ Failed for TDR {tdr_number}: {e}")

def main():
    login()
    df = pd.read_excel(TDR_LIST_FILE)  # Ensure it has column named 'TDR'
    for tdr in df['TDR']:
        search_tdr(tdr)
        download_boq(str(tdr))
    driver.quit()
    print("✅ All done.")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import os

def download_boq(tender_id):
    base_url = "https://eprocure.gov.in/eprocure/app"
    search_url = f"{base_url}?page=FrontEndAdvancedSearchResult&service=page"

    # Start a session
    session = requests.Session()

    # Search for the tender
    search_params = {
        'tenderId': tender_id
    }
    response = session.get(search_url, params=search_params)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the tender details link
    tender_links = soup.find_all('a', href=True)
    tender_detail_link = None
    for link in tender_links:
        if tender_id in link.text:
            tender_detail_link = link['href']
            break

    if not tender_detail_link:
        print(f"Tender ID {tender_id} not found.")
        return

    # Access the tender details page
    tender_detail_url = f"{base_url}{tender_detail_link}"
    response = session.get(tender_detail_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the BOQ file link
    boq_link = None
    for link in soup.find_all('a', href=True):
        if 'BOQ' in link.text and link['href'].endswith('.xls'):
            boq_link = link['href']
            break

    if not boq_link:
        print(f"BOQ file not found for Tender ID {tender_id}.")
        return

    # Download the BOQ file
    boq_url = f"{base_url}{boq_link}"
    response = session.get(boq_url)

    # Save the BOQ file
    filename = boq_link.split('/')[-1]
    with open(filename, 'wb') as file:
        file.write(response.content)

    print(f"BOQ file downloaded: {filename}")

# Example usage
tender_id = "2025_ARD_850268_1"
            
# Replace with your Tender ID
download_boq(tender_id)
```

```python
import requests
from bs4 import BeautifulSoup
import os

def download_boq(tender_id, save_dir='boq_files'):
    base_url = 'https://eprocure.gov.in/eprocure/app'
    search_url = f"{base_url}?component=$DirectLink_0&page=FrontEndAdvancedSearchResult&service=direct&sp={tender_id}"

    headers = {
        'User-Agent': 'Mozilla/5.0'
    }

    # Create directory to save BOQ files
    os.makedirs(save_dir, exist_ok=True)

    try:
        # Fetch the tender page
        response = requests.get(search_url, headers=headers, verify=True)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all links ending with .xls (BOQ files)
        for link in soup.find_all('a', href=True):
            if link['href'].endswith('.xls'):
                boq_url = base_url + link['href']
                boq_response = requests.get(boq_url, headers=headers, verify=True)
                boq_response.raise_for_status()

                # Save the BOQ file
                file_name = link['href'].split('/')[-1]
                file_path = os.path.join(save_dir, file_name)
                with open(file_path, 'wb') as file:
                    file.write(boq_response.content)

                print(f"BOQ file downloaded and saved as: {file_path}")
                return

        print("BOQ file not found for the given Tender ID.")

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")

# Example usage
download_boq('2025_ARD_850268_1')
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import os
import urllib.request

# Load Tender IDs from Excel
excel_path = "all_data.xlsx"  # Update if file is elsewhere
df = pd.read_excel(excel_path)
tender_ids = df["TenderId"].dropna().unique()

# Setup Chrome Driver
download_dir = os.path.join(os.getcwd(), "BOQ_WB_Files")
os.makedirs(download_dir, exist_ok=True)

chrome_options = Options()
chrome_options.add_argument("--start-maximized")
prefs = {"download.default_directory": download_dir}
chrome_options.add_experimental_option("prefs", prefs)

driver = webdriver.Chrome(options=chrome_options)

# Navigate to Advance Search Page
driver.get("https://wbtenders.gov.in/nicgep/app?page=FrontEndAdvancedSearch&amp;service=page")
time.sleep(3)

print("📦 Starting BOQ download from WB Portal...")

for tender_id in tender_ids:
    try:
        print(f"🔍 Checking: {tender_id}")
        driver.get("https://wbtenders.gov.in/nicgep/app?page=FrontEndAdvancedSearch&service=page")
        time.sleep(2)

        # Fill Tender ID
        search_input = driver.find_element(By.ID, "tenderId")
        search_input.clear()
        search_input.send_keys(tender_id)

        # Click Search
        driver.find_element(By.ID, "submit").click()
        time.sleep(3)

        # Click Tender Title link
        rows = driver.find_elements(By.CSS_SELECTOR, "table[class*='table'] tbody tr")
        if not rows:
            print(f"❌ Tender ID not found: {tender_id}")
            continue

        # Click the first tender title link
        rows[0].find_element(By.TAG_NAME, "a").click()
        time.sleep(3)

        # Find and click BOQ link
        links = driver.find_elements(By.TAG_NAME, "a")
        boq_found = False
        for link in links:
            href = link.get_attribute("href")
            text = link.text.strip()
            if "BOQ" in text and href.endswith(".xls"):
                boq_url = href
                filename = f"{tender_id}_BOQ.xls"
                save_path = os.path.join(download_dir, filename)
                urllib.request.urlretrieve(boq_url, save_path)
                print(f"✅ BOQ downloaded: {filename}")
                boq_found = True
                break

        if not boq_found:
            print(f"❌ BOQ not found for: {tender_id}")

    except Exception as e:
        print(f"⚠️ Error on {tender_id}: {e}")

print(f"\n🎉 Done. BOQ files saved in: {download_dir}")
driver.quit()
```

```python
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Load Tender IDs
df = pd.read_excel("all_data.xlsx")
tender_ids = df['TenderId'].dropna().unique().tolist()

# Setup download path
download_dir = os.path.join(os.getcwd(), "BOQ_WB_Files")
os.makedirs(download_dir, exist_ok=True)

# Chrome options
chrome_options = Options()
chrome_options.add_experimental_option("prefs", {
    "download.default_directory": download_dir,
    "download.prompt_for_download": False,
    "plugins.always_open_pdf_externally": True
})
chrome_options.add_argument("--start-maximized")

# Launch Chrome
driver = webdriver.Chrome(options=chrome_options)
driver.get("https://wbtenders.gov.in/nicgep/app")

# CAPTCHA prompt
input("🔐 Solve the CAPTCHA manually in browser, then press Enter here...")

print("📦 Starting BOQ download from WB eTender Portal...")

for tender_id in tender_ids:
    try:
        driver.get("https://wbtenders.gov.in/nicgep/app?page=FrontEndAdvancedSearchURL&service=page")

        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "tenderId"))
        )

        search_box = driver.find_element(By.ID, "tenderId")
        search_box.clear()
        search_box.send_keys(tender_id)

        driver.find_element(By.ID, "searchBtn").click()

        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "table"))
        )

        links = driver.find_elements(By.XPATH, "//a[contains(text(),'BOQ')]")
        if links:
            for link in links:
                file_url = link.get_attribute("href")
                driver.execute_script("window.open(arguments[0]);", file_url)
                time.sleep(3)
                driver.switch_to.window(driver.window_handles[-1])
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
            print(f"✅ BOQ downloaded for: {tender_id}")
        else:
            print(f"❌ BOQ not found for: {tender_id}")

    except Exception as e:
        print(f"⚠️ Error with {tender_id}: {e}")

print(f"🎉 Done. BOQ files saved in: {download_dir}")
driver.quit()
```
