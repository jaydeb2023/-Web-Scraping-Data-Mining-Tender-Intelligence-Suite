```python
import os
import shutil
import pandas as pd

# Folder containing all extracted BOQ files
base_dir = r'di_PIPE'
output_dir = os.path.join(base_dir, 'Filtered_BOQs_DI')
os.makedirs(output_dir, exist_ok=True)

# Keywords to search for
keywords = ['di pipe', 'ductile iron pipe']

# Traverse all subfolders and files
for root, _, files in os.walk(base_dir):
    for file in files:
        if file.lower().endswith(('.xls', '.xlsx')):
            file_path = os.path.join(root, file)
            try:
                # Read Excel file
                df = pd.read_excel(file_path, engine='openpyxl' if file.endswith('.xlsx') else 'xlrd')
                df.columns = df.columns.map(str)

                # Check for keyword match
                if any(
                    df[col].astype(str).str.lower().str.contains('|'.join(keywords), na=False).any()
                    for col in df.select_dtypes(include='object').columns
                ):
                    shutil.copy(file_path, output_dir)
                    print(f"✅ DI pipe found in: {file_path}")
            except Exception as e:
                print(f"❌ Skipped {file_path} due to error: {e}")

print(f"\n🏁 Done! Filtered files are in: {output_dir}")
```
