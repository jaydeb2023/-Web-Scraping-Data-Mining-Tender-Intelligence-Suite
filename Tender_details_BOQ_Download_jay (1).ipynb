```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
import pandas as pd

# Load Excel file with TDR numbers
df = pd.read_excel("MP_pipe_data.xlsx")
tdr_list = df["TDR"].astype(str).tolist()

# Set download directory
download_dir = os.path.join(os.getcwd(), "BOQ_Downloads")
os.makedirs(download_dir, exist_ok=True)

# Setup Chrome options
options = webdriver.ChromeOptions()
prefs = {"download.default_directory": download_dir}
options.add_experimental_option("prefs", prefs)

driver = webdriver.Chrome(options=options)

# Step 1: Login manually
driver.get("https://www.tenderdetail.com/Account/LogOn")
input("🔐 Log in manually, then press ENTER to continue...")

# Step 2: Process each TDR number
for tdr in tdr_list:
    print(f"🔎 Searching for TDR: {tdr}")
    search_url = f"https://www.tenderdetail.com/registeruser/searchtenders?ServiceType=1&nm={tdr}"
    driver.get(search_url)
    time.sleep(3)

    # Step 3: Check if TDR result is valid
    try:
        tender_brief = driver.find_element(By.CSS_SELECTOR, "#tenderbreif > span").text
        if tdr not in tender_brief:
            print(f"⚠️ TDR mismatch or not found: {tdr}")
            continue
    except:
        print(f"❌ No result for TDR: {tdr}")
        continue

    # Step 4: Click on "View Notice"
    try:
        view_notice = driver.find_element(By.CSS_SELECTOR, ".viewnotice")
        driver.execute_script("arguments[0].click();", view_notice)
        time.sleep(3)
    except:
        print(f"❌ Could not click 'View Notice' for TDR: {tdr}")
        continue

    # Step 5: Switch to new tab
    driver.switch_to.window(driver.window_handles[-1])
    time.sleep(3)

    # Step 6: Look for PDF/XLS download links
    all_links = driver.find_elements(By.TAG_NAME, "a")
    downloaded = False
    for link in all_links:
        href = link.get_attribute("href")
        if href and (".xls" in href or ".xlsx" in href or "BOQ" in href):
            driver.get(href)
            print(f"📥 Downloading: {href}")
            downloaded = True
            time.sleep(3)
            break

    if not downloaded:
        print(f"❌ BOQ not found for TDR: {tdr}")

    # Step 7: Close tab and return to main
    if len(driver.window_handles) > 1:
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    time.sleep(2)

print("✅ All done!")
driver.quit()
```
