```python
df = pd.read_excel("BOQ_1833226.xls", skiprows=10)
```

```python
rate_col = next((col for col in df.columns if "Estimated Rate" in col), None)
```

```python
df = df[df['Units'].astype(str).str.strip().str.lower() == 'mtr.']
```

```python
def extract_class(desc):
    desc = str(desc).upper()
    if 'K-9' in desc or 'K9' in desc:
        return 'K9'
    elif 'K-7' in desc or 'K7' in desc:
        return 'K7'
    elif 'M.S' in desc or 'MS' in desc:
        return 'M.S'
    elif 'DI' in desc:
        return 'DI'
    elif 'CI' in desc:
        return 'CI'
    else:
        return 'Other'
```

```python
def extract_dia(desc):
    desc = str(desc)
    match = re.search(r'(\d{2,4})\s*mm', desc)
    return match.group(1) if match else ''
```

```python
df['class'] = df['Item Description'].apply(extract_class)
df['DIA'] = df['Item Description'].apply(extract_dia)
```

```python
df = df.rename(columns={rate_col: 'estimate rate'})
```
