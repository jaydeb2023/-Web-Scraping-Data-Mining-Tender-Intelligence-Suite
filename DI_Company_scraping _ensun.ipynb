```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Setup Selenium
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
# options.add_argument('--headless')  # Optional for background run
driver = webdriver.Chrome(options=options)

pipe_type = "Ductile Iron Pipe"
all_data = []

# Step 1: Scrape profile links from search pages
profile_links = []

for page in range(1, 21):
    print(f"\nScraping page {page} for '{pipe_type}'...")

    url = f"https://ensun.io/search?threshold=VERY_LOW&q={pipe_type.replace(' ', '%20')}&page={page}&locations=United%20States%2Cnull%2Cnull&categories=MANUFACTURER"
    driver.get(url)
    time.sleep(4)

    listings = driver.find_elements(By.CLASS_NAME, "search-result-link")

    if not listings:
        print("No more results found, stopping.")
        break

    for item in listings:
        try:
            company_link = item.get_attribute("href")
            profile_links.append(company_link)
        except:
            continue

print(f"\nFound {len(profile_links)} company profile links. Now extracting details...\n")

# Step 2: Visit each profile and extract details
for link in profile_links:
    driver.get(link)
    time.sleep(4)

    try:
        company_name = driver.find_element(By.XPATH, '//h1').text.strip()
    except:
        company_name = "Not Found"

    try:
        country = driver.find_element(By.XPATH, "//span[contains(text(),'Country')]/following-sibling::span").text.strip()
    except:
        country = "Not Found"

    try:
        industry = driver.find_element(By.XPATH, "//span[contains(text(),'Industry')]/following-sibling::span").text.strip()
    except:
        industry = "Not Found"

    try:
        employees = driver.find_element(By.XPATH, "//span[contains(text(),'Employees')]/following-sibling::span").text.strip()
    except:
        employees = "Not Found"

    try:
        founded = driver.find_element(By.XPATH, "//span[contains(text(),'Founded')]/following-sibling::span").text.strip()
    except:
        founded = "Not Found"

    all_data.append({
        "Company Name": company_name,
        "Country": country,
        "Industry": industry,
        "Employee Count": employees,
        "Year Founded": founded,
        "Profile Link": link
    })

driver.quit()

# Save to Excel
df = pd.DataFrame(all_data)
output_file = "Ensun_Ductile_Iron_Pipe_Company_Details.xlsx"
df.to_excel(output_file, index=False)
print(f"\n✅ Data saved to '{output_file}'")
```
